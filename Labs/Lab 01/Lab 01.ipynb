```python
def sundaes():
    flavors = ["vanilla", "chocolate", "strawberry", "pistacchio"]
    sauces = ["caramel", "butterscotch", "chocolate"]
    count = 0
    for flavor in flavors:
        for sauce in sauces:
            print(flavor + " ice cream sundae with " + sauce + " sauce")
            count += 1
    return count

print("Total combinations: ", sundaes())
```

```python
def triangle():
    value = 1
    row = 1
    while row <= 10:
        column = 1
        while column <= row:
            if column != row:
                print(value, ' ', sep = '', end = '')
            else:
                print(value)
            value = value + 1
            column = column + 1
        row = row + 1

triangle()
```

```python
def cube(n):
    return n ** 3

def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def count_pattern(pattern, lst):
    count = 0
    p_len = len(pattern)

    for i in range(len(lst) - p_len + 1):
        if(lst[i:i + p_len] == pattern):
            count += 1
    return count

def multiplication_table(n):
    for i in range(1,11):
        print(n, "x", i, "=", n * i)

def calculator(a, b, op):
    if op == '-':
        return a - b
    elif op == '+':
        return a + b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return "Invalid operator"

def sort_sentence(sentence):
    words = sentence.lower().split()
    words.sort()
    return ' '.join(words)

print("Cube:", cube(3))
print("Factorial:", factorial(5))
print("Pattern Count:", count_pattern(('a','b'), ('a','b','c','e','b','a','b','f')))
print("Sorted Sentence:", sort_sentence("This is the first sentence"))

print("Calculator (+):", calculator(5, 3, '+'))
print("Calculator (-):", calculator(5, 3, '-'))
print("Calculator (*):", calculator(5, 3, '*'))
print("Calculator (/):", calculator(6, 3, '/'))

print("\nMultiplication Table:")
multiplication_table(5)
```

```python
class Roman:
    def int_to_roman(self, num):
        values = [1000,900,500,400,100,90,50,40,10,9,5,4,1]
        symbols = ["M","CM","D","CD","C","XC","L","XL","X","IX","V","IV","I"]
        result = ""
        for i in range(len(values)):
            while num >= values[i]:
                result += symbols[i]
                num -= values[i]
        return result

r = Roman()
print(r.int_to_roman(354))
```

```python
class TwoSum:
    def find_indices(self, numbers, target):
        seen = {}
        result = []
        for i, num in enumerate(numbers):
            diff = target - num
            if diff in seen:
                for j in seen[diff]:
                    result.append((j,i))
            if num in seen:
                seen[num].append(i)
            else:
                seen[num] = [i]
        return result
obj = TwoSum()
print(obj.find_indices([10,20,10,40,50,60,70], 50))
```

```python
class ThreeSum:
    def find_triplets(self, arr):
        arr.sort()
        result = []
        n = len(arr)

        for i in range(n - 2):
            if i > 0 and arr[i] == arr[i - 1]:
                continue  # avoid duplicates

            left, right = i + 1, n - 1

            while left < right:
                total = arr[i] + arr[left] + arr[right]

                if total == 0:
                    result.append([arr[i], arr[left], arr[right]])
                    left += 1
                    right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result

obj = ThreeSum()
print(obj.find_triplets( [-25, -10, -7, -3, 2, 4, 8, 10]))
```

```python
class MatrixAddition:
    def add(self, A, B):
        n = len(A)
        C = []

        for i in range(n):
            row = []
            for j in range(n):
                row.append(A[i][j] + B[i][j])
            C.append(row)

        return C

A = [[1, 2],
     [3, 4]]

B = [[5, 6],
     [7, 8]]

obj = MatrixAddition()
result = obj.add(A, B)

print("Addition Result:")
for row in result:
    print(row)
```

```python
class MatrixMultiplication:
    def multiply(self, A, B):
        r1 = len(A)
        c1 = len(A[0])
        r2 = len(B)
        c2 = len(B[0])
        C = [[0 for _ in range(c2)] for _ in range(r1)]
        for i in range(r1):
            for j in range(c2):
                for k in range(c1):
                    C[i][j] += A[i][k] * B[k][j]
        return C

A = [
    [1, 2, 3],
    [4, 5, 6]
]
B = [
    [7, 8],
    [9, 10],
    [11, 12]
]
obj = MatrixMultiplication()
result = obj.multiply(A, B)
for row in result:
    print(row)
```

```python
def calculator(a, b, op='+', output_format='float'):
    try:
        if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
            raise TypeError("Inputs must be numbers")

        if op not in ['+', '-', '*', '/']:
            raise ValueError("Invalid operation")

        if op == '+':
            result = a + b
        elif op == '-':
            result = a - b
        elif op == '*':
            result = a * b
        elif op == '/':
            if b == 0:
                raise ZeroDivisionError("Cannot divide by zero")
            result = a / b

        if output_format == 'int':
            return round(result)
        elif output_format == 'float':
            return float(result)
        else:
            raise ValueError("Invalid output format")

    except (TypeError, ValueError, ZeroDivisionError) as e:
        return "Error:", str(e)

print(calculator(5, 2))
print(calculator(5, 2, '*', 'int'))
print(calculator(5, 2, '/', 'float'))
print(calculator(5, 0, '/', 'float'))
print(calculator(5, 2, '%', 'float'))
print(calculator("5", 2, '+', 'float'))
```

```python
class Numbers:
    MULTIPLIER = 2
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def add(self):
        return self.x + self.y

    @classmethod
    def multiply(cls, a):
        return a * cls.MULTIPLIER

    @staticmethod
    def subtract(b, c):
        return b - c

    @property
    def value(self):
        return (self.x, self.y)

    @value.setter
    def value(self, new_values):
        self.x, self.y = new_values

    @value.deleter
    def value(self):
        self.x = None
        self.y = None

obj = Numbers(5, 10)
print(obj.add())
print(Numbers.multiply(3))
print(Numbers.subtract(10, 4))
print(obj.value)
obj.value = (7, 8)
print(obj.value)
del obj.value
print(obj.value)
obj.value = (1,2)
print(obj.value)
```
